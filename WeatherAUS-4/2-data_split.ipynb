```python
import pandas as pd
```

```python
data = pd.read_csv("data/weatherAUS.csv")
```

```python
data.shape
```

```python
def current_future_split(data, future_after):
    years = pd.to_datetime(data['Date'])
    future = years.dt.year > future_after

    return data[~future], data[future]
```

```python
current_data, future_data = current_future_split(data, 2015)

print(current_data.shape)
print(future_data.shape)
```

```python
current_data.info()
```

```python
future_data.info()
```

```python
current_data.to_csv("data/current_data.csv", index=False)
future_data.to_csv("data/future_data.csv", index=False)
```
